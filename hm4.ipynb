# Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости на основе матрицы корреляции.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
matrix = df.corr(numeric_only=True)
matrix = np.round(matrix, 1)
matrix[np.abs(matrix) < 0.3] = 0
matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_lot'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

### Цена домов не слишком зависит от общей площади, но зависит от жилой (чем больше жилая площадь, тем выше стоиомсть домов)

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

### чем выше grade, тем выше цена дома

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Этажи')
plt.title('Распределение цен в зависимости от этажности');
```

### выше всего цена у домов с количество этажей 2,5 и 2

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Вид')
plt.title('Распределение цен в зависимости от вида из дома');
```

### выше всего цена у домов с видом 4 и 3
